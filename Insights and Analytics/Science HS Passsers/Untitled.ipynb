#### Libraries/Packages

```python
import pdfplumber
import pandas as pd
```

#### Data

```python
FILE_PATH = r"./RM No. 277 s. 2026 - Released of the List of Qualifiers Eligible for Enrollment to NCR Science High Schools, SY 2026 - 2027.pdf"
```

```python
# Open PDF and iterate through all pages to extract tables
with pdfplumber.open(FILE_PATH) as pdf:
    data = [page.extract_table() for page in pdf.pages if page.extract_table()]
```

```python
DFS = [pd.DataFrame(table) for table in data]
```

```python
DF_COMBINED = pd.concat(DFS, ignore_index=True)
```

```python
DF_COMBINED = DF_COMBINED.drop_duplicates(keep="first")
```
