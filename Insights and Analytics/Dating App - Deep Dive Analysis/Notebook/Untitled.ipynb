```python
import pandas as pd
import numpy as np
```

```python
FILEPATH = "../Data/"
```

```python
df = pd.read_csv(FILEPATH + "user_data.csv")

# Remove index column
df.drop(columns=df.columns[0], inplace=True)
```

```python
df.head()
```

```python
df["gender"].value_counts() / df.shape[0]
```

```python
df.groupby(["gender"])[["likes","matches"]].mean()
```

```python
df["attractiveness_tier"] = [round(np.floor(x*100/10)+1,0) for x in df["attractiveness"]]
```

```python
df.groupby(["gender","attractiveness_tier"]).agg({
                                                 "likes": "mean",
                                                 "matches": "mean",
                                                 "gender": "count"
                                                })
```
