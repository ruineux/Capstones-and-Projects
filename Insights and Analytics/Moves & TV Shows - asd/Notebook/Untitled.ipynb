### Name/Basics
- nconst (string) - alphanumeric unique identifier of the name/person
- primaryName (string)– name by which the person is most often credited
- birthYear – in YYYY format
- deathYear – in YYYY format if applicable, else '\N'
- primaryProfession (array of strings)– the top-3 professions of the person
- knownForTitles (array of tconsts) – titles the person is known for

```python
imdb_namebasics = pd.read_table("datasource/"+ "name.basics.tsv.gz", compression="gzip")

# Replacing Not applicable values with null
imdb_namebasics.replace("\\N", np.nan, inplace=True)

# Removing no identified primaryProfession and knowForTitles
imdb_namebasics = \
imdb_namebasics[~(imdb_namebasics["primaryProfession"].isnull() | imdb_namebasics["knownForTitles"].isnull())].reset_index(drop=True)
```

```python
imdb_namebasics
```

#### Text to Columns: primaryProfession

```python
imdb_namebasics = \
imdb_namebasics.drop(columns=["primaryProfession"]).join(imdb_namebasics['primaryProfession'].str.split(',',expand=True).add_prefix('primaryProfession_').fillna(np.nan))
```

#### Text to Columns: knownForTitles

```python
imdb_namebasics = \
imdb_namebasics.drop(columns=["knownForTitles"]).join(imdb_namebasics['knownForTitles'].str.split(',',expand=True).add_prefix('knownForTitles_').fillna(np.nan))
```
