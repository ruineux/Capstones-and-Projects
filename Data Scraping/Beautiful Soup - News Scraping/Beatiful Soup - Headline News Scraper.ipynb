### Get all headlines

```python
def inquirer_headlines(html_data):
    headline_source = []
    headline_title = []
    headline_link = []

    for headline in html_data:
        headline_source.extend([[headline_source.append(headline.find("div", class_="flx-m-cat").find("span").text),
         headline_source.append(headline.find("div", class_="flx-m-head").text),
         headline_source.append(headline.find("a", href=True)["href"])]])

    return headline_source
```
